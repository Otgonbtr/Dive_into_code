```python
###### Problem 1 ###########
# Data acquisition

from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

data = load_iris()
X = pd.DataFrame(data=data.data, columns=data.feature_names)
Y = pd.DataFrame(data=data.target, columns=['Species'])
#print(X)
#print(Y)
```

```python
###### Problem 2 #############
# Combining Data
X = pd.concat([X,Y], axis=1)
print(X.head())
```

```python
######## Problem 3 ###########
# Checking Data
print(X.head(4))
print(X.info())
print(X.describe())
print('4th sample: {}'.format(X.loc[4]))
print('total number of samples for each label: {}'.format(X.value_counts('Species')))
count = np.count_nonzero(X.isnull())
print('There is {} missing values.'.format(count))
print('sum is: {}'.format(X.sum()))
```

```python
######## Problem 6 #########
# Creating a diagram
import matplotlib.pyplot as plt
labels = ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)','Species']
fig, ax = plt.subplots()
ax.pie(X.sum(), labels = labels, autopct='%1.0f%%')
plt.show()
```

```python
fig,ax = plt.subplots()
labels = ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
ax.boxplot((X.iloc[:,0],X.iloc[:,1],X.iloc[:,2],X.iloc[:,3]),labels=labels)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.violinplot((X.iloc[:,0],X.iloc[:,1],X.iloc[:,2],X.iloc[:,3]))
plt.show()
```

```python
####### Problem 7 ###########
# Confirmation of the relationship between features
import seaborn as sns
sns.scatterplot(X['sepal length (cm)'],X['sepal width (cm)'],hue=X['Species'])
plt.show

sns.pairplot(X, vars=["sepal width (cm)", "sepal length (cm)"], hue='Species', kind='scatter')
plt.show()

sns.pairplot(X, vars=["petal width (cm)", "petal length (cm)"], hue="Species", kind='scatter')
plt.show()

sns.pairplot(X, vars=["sepal width (cm)", "petal width (cm)"], hue="Species", kind='scatter')
plt.show()

sns.pairplot(X, vars=["sepal length (cm)", "petal length (cm)"], hue="Species", kind='scatter')
plt.show()

sns.pairplot(X, hue = 'Species', kind='scatter')
plt.show()

corr = X.corr(method = 'pearson')
print(corr)

sns.heatmap(corr)
plt.title('Correlation')
plt.show()
```
