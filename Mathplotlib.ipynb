# Stacked bar chart

Багцалсан бар хүснэгт нь багананд багцлагдсан олон элементийн өгөгдлийг харуулдаг барын хүснэгт юм.
Элемент тус бүрийг өнгөөр кодчилсноор та нийт болон бүрэлдсэн элементийн хэмжээг харж уншиж болно. Мөн элемент бүрийн өсөлтийн хэмжээг сар сараас сар хүртэл төсөөлөхөд ихэвчлэн ашигладаг.

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

x = [1,2,3]
y1 = [4,12,5]
y2 = [8,2,4]
y_total = np.add(y1,y2)
labels = ['Apple', 'Banana', 'Orange']

ax.bar(x,y_total, tick_label=labels, label='y2')
ax.bar(x,y1, label='y1')
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

x = [1,2,3]
y1 = [4,12,5]
y2 = [8,2,4]
y3 = [2,6,8]
y1_y2_total = np.add(y1,y2)
y_total = np.add(y1_y2_total,y3)
labels = ['Apple', 'Banana', 'Orange']

ax.bar(x,y_total, tick_label=labels, label='y3')
ax.bar(x,y1_y2_total, tick_label=labels, label='y2')
ax.bar(x,y1, label='y1')
ax.legend()
plt.show()
```

## Histogram

Гистограм гэдэг нь бар хүснэгттэй төстэй схемд давтамжийн хүснэгтийг төлөөлөх график юм.
Давтамжийн хүснэгт гэдэг нь өгөгдлийг дур зоргоороо хуваасан хүснэгт бөгөөд төрөл бүрийн дотор байгаа өгөгдлийн тоог хүснэгтэд хураангуйлдаг.
Гистограм нь босоо тэнхлэг дээрх өгөгдлийн class (range) болон босоо тэнхлэг дээрх тухайн class-д байгаа өгөгдлийн тоог авдаг.Гистограммд хуримтлагдсан өгөгдлийг дүрсжүүснээр үйлчилгээ эрхлэгчдийн насны бүлэг, өдөр тутмын зочлох болон вэбсайтын PV-ийн тоо, бизнесийн урсгалын саад бэрхшээл зэрэг бүх сэдвүүдийн зургийг бүхэлд нь шууд ойлгох боломжтой.

- yamagata - Мэдээллийн хувьсал харьцангуй бага байгааг илтгэнэ.
- Pointed Yamagata - Өгөгдлийн хувьсал бага байгааг илтгэнэ.
- Gentle Yamagata - Мэдээллийн өндөр хувьсалыг илтгэдэг.
- Multiple mountains - олон төрлийн өөрчлөлт байгааг илтгэнэ.
- 	Distorted Yamagata - Мэдээлэлд тал тохой татахыг илтгэдэг.

# Box plot

Box plot нь дээд, доод талд нь хязгаарлалт гардаг дөрвөлжин хайрцаг бөгөөд өгөгдлийн хувьсалыг харуулахад ашиглагддаг.
Та гистограмм дахь өгөгдлийн хувьсалыг харж болно, гэхдээ та box plot ашиглан олон өгөгдлийн хувьсалыг харьцуулж болно.
Box plots нь квартил ашиглан өгөгдөл тархахыг харуулдаг. Энэ нь квартил дахь өгөгдлийн хувьсалыг харуулдаг.
Хамгийн бага утга тооцоход нийт тоонуудын 25%-д тохирох утга нь эхний квартил, дунд утга нь дундаж (50%), 75% -д хамаарах утга нь гурав дахь квартил юм.Үнэт цаасны болон валютын ханшийн чиг хандлагыг харахын тулд box plots-ийг түгээмэл ашигладаг.


## Drawing a box plot

```python
import numpy as np
x1 = np.random.normal(0,10,1000)
x2 = np.random.normal(0,15,1000)
x3 = np.random.normal(0,20,1000)

import matplotlib.pyplot as plt
fig,ax = plt.subplots()
labels=['x1','x2','x3']
ax.boxplot((x1,x2,x3),labels=labels)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.boxplot((x1,x2,x3),labels=labels , vert= False)
plt.show()
```

- box plot гэдэг нь квартил ашиглан өгөгдлийн хувьсах хэмжээг харуулсан график юм. Олон өгөгдлийн хувьсалыг харьцуулахыг хүссэн үед тохиромжтой.
- boxplot method-ийг ашиглан  boxplot-ийг зурна.

## Drawing a pie chart

```python
import pandas as pd
df = pd.DataFrame([['example1',5],
                   ['example2',15],
                   ['example3',25],
                   ['example4',50],
                   ['example5',1],
                   ['example6',4]], columns = ["News","arg_results(%)"])
df
```

```python
import matplotlib.pyplot as plt
labels = df.loc[:,"News"]
rate = df.loc[:,"arg_results(%)"]

fig,ax = plt.subplots()
ax.pie(rate, labels=labels)
ax.axis('equal')
plt.show()
```

ax.axis('equal') гэдэг нь зурах графикийн тал харьцааг (графикийн өндөр ба өргөний харьцаа) тогтоох арга юм.

## Pie chart formating

```python
labels = df.loc[:,"News"]
rate = df.loc[:,"arg_results(%)"]

fig,ax = plt.subplots()
# startangle=90
ax.pie(rate, labels=labels, startangle=90)
ax.axis('equal')
plt.show()
```

```python
labels = df.loc[:,"News"]
rate = df.loc[:,"arg_results(%)"]

fig,ax = plt.subplots()
# counterclock=False
ax.pie(rate, labels=labels, startangle=90 , counterclock=False)
ax.axis('equal')
plt.show()
```

```python
labels = df.loc[:,"News"]
rate = df.loc[:,"arg_results(%)"]

fig,ax = plt.subplots()
# shadow=True
ax.pie(rate, labels=labels, startangle=90 , counterclock=False, shadow=True)
ax.axis('equal')
plt.show()
```
