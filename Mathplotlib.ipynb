```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

x = [1,2,3]
y1 = [4,12,5]
y2 = [8,2,4]
y_total = np.add(y1,y2)
labels = ['Apple', 'Banana', 'Orange']

ax.bar(x,y_total, tick_label=labels, label='y2')
ax.bar(x,y1, label='y1')
ax.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

x = [1,2,3]
y1 = [4,12,5]
y2 = [8,2,4]
y3 = [2,6,8]
y1_y2_total = np.add(y1,y2)
y_total = np.add(y1_y2_total,y3)
labels = ['Apple', 'Banana', 'Orange']

ax.bar(x,y_total, tick_label=labels, label='y3')
ax.bar(x,y1_y2_total, tick_label=labels, label='y2')
ax.bar(x,y1, label='y1')
ax.legend()
plt.show()
```

## Drawing a pie chart

```python
import pandas as pd
df = pd.DataFrame([['example1',5],
                   ['example2',15],
                   ['example3',25],
                   ['example4',50],
                   ['example5',1],
                   ['example6',4]], columns = ["News","arg_results(%)"])
df
```

```python
import matplotlib.pyplot as plt
labels = df.loc[:,"News"]
rate = df.loc[:,"arg_results(%)"]

fig,ax = plt.subplots()
ax.pie(rate, labels=labels)
ax.axis('equal')
plt.show()
```

ax.axis('equal')       (    )   .

## Pie chart formating

```python
labels = df.loc[:,"News"]
rate = df.loc[:,"arg_results(%)"]

fig,ax = plt.subplots()
# startangle=90
ax.pie(rate, labels=labels, startangle=90)
ax.axis('equal')
plt.show()
```

```python
labels = df.loc[:,"News"]
rate = df.loc[:,"arg_results(%)"]

fig,ax = plt.subplots()
# counterclock=False
ax.pie(rate, labels=labels, startangle=90 , counterclock=False)
ax.axis('equal')
plt.show()
```

```python
labels = df.loc[:,"News"]
rate = df.loc[:,"arg_results(%)"]

fig,ax = plt.subplots()
# shadow=True
ax.pie(rate, labels=labels, startangle=90 , counterclock=False, shadow=True)
ax.axis('equal')
plt.show()
```

```python
labels = df.loc[:,"News"]
rate = df.loc[:,"arg_results(%)"]

fig,ax = plt.subplots()
# counterclock=False
ax.pie(rate, labels=labels, startangle=90 , counterclock=False, autopct='%1.0f%%')
ax.axis('equal')
plt.show()
```

```python
labels = df.loc[:,"News"]
rate = df.loc[:,"arg_results(%)"]

fig,ax = plt.subplots()
# explode 
explode=[0.2,0,0,0,0,0]
ax.pie(rate, labels=labels, startangle=90 , counterclock=False, autopct='%1.0f%%', explode=explode)
ax.axis('equal')
plt.show()
```

- explode-oor todorhoi hemjeenii utgiig salgaj avch haruulj bolno gehde zuvhun ehnii hesgiig ni l salgaj haruulna.
