```python
import numpy as np
import matplotlib.pyplot as plt

###### Problem 1 #########
# Creation of random numbers

np.random.seed(0)
mean = [-3,0]
mean2 = [0,-3]
cov = ([1.0,0.8],[0.8,1.0])

rand_data1 = np.random.multivariate_normal(mean,cov,size=500)
rand_data2 = np.random.multivariate_normal(mean2,cov,size=500)
###### Problem 2 #########
# Visualization by scatter plot
fig ,ax = plt.subplots()
ax.scatter(rand_data1[:,0],rand_data1[:,1])
fig.suptitle('scatter')
ax.axis('equal')
plt.show()
```

```python
###### Problem 3 #########
# Visualization by histogram

fig , ax = plt.subplots()
ax.set_title('Histogram of x1')
ax.set_xlabel('x1')
ax.set_ylabel('frequency')
plt.xlim((-5,3))
ax.hist(rand_data1[:,0],bins=60)

fig , ax = plt.subplots()
ax.set_title('Histogram of x2')
ax.set_xlabel('x2')
ax.set_ylabel('frequency')
plt.xlim(-5,3)
ax.hist(rand_data1[:,1],bins=60)
plt.show()
```

```python
######## Problem 4 ##########
# Addition of data
fig, ax = plt.subplots()
ax.scatter(rand_data1[:,0],rand_data1[:,1],label=0)
ax.scatter(rand_data2[:,0],rand_data2[:,1],label=1)
plt.title('scatter')
plt.axis('equal')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.show()
```

```python
########### Problem 5 ###########
# Data combination

#use np.concatenate()
combined_data = np.concatenate((rand_data1,rand_data2))
print('Combined data by np.concatenate shape {}'.format(combined_data1.shape))

#use np.vstask
combined_data = np.vstack((rand_data1,rand_data2))
print('Combined data by np.vstack shape {}'.format(combined_data.shape))
```

```python
######### Problem 6 ##########
# Labeling

label = np.empty((1000,1))
label[0:500] = 0
label[500:1000] = 1

combined_data = np.append(combined_data,label, axis = 1)
print(combined_data.shape)
print(combined_data)
```
